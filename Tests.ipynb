```python
import json
from equation_parser import *
```

```python
rules = ('1+2', 
         '1/1',
         '1*3',
         '4-2',
         'col_1 + col_2',
         'col_1 - col_2',
         'col_1 * col_2',
         'col_1 / col_2',
         '(1.2*col_1) + 2*col_2 / 100', 
         '1000*col_1**2',
         '+ 1 / sin(col_1)',
         'sin(col_1)**2 + cos(col_1)**2',
         'atan(col_1 + 1)',
         'arctan(col_1 + 1)',
         'arctan2(col_1 + 1)',
         '- 1 * col_1+2 -(3*col_2)/(col_1) ', 
         'sin(col_2+2) -(col_1 + 1)',
         'abs(col_1 - 2)',
         'exp(-col_1)',
         'sqrt(col_1)',
         'b100(col_1)',
         'delta(col_2)',
         'b100(col_1 + col_2)')
```

```python
results = []
for rule in rules:
    tree = from_string(rule)
    serialized = tree.serialize()
    results.append({'input': rule, 'output': serialized})
```

```python
with open('./tests/test.json', mode='w') as f:
    json.dump(results, f, indent=4)
```
