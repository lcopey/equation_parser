```python
    ast.Add: Operators.ADD,
    ast.Sub: Operators.SUB,
    ast.Mult: Operators.MUL,
    ast.Div: Operators.DIV,
    ast.Pow: Operators.POW,
    ast.USub: Operators.NEG,
    ast.UAdd: Operators.ID,
    'sin': Operators.SIN,
    'cos': Operators.COS,
    'tan': Operators.TAN,
    'atan': Operators.ATAN,
    'arctan': Operators.ATAN,
    'arctan2': Operators.ATAN,
    'abs': Operators.ABS
```

```python
rules = ('1+2', 
         '(1.2*col_1) + 2*col_2 / 100', 
         '1000*col_1**2',
         '+ 1 / sin(col_1)',
         'sin(col_1)**2 + cos(col_1)**2',
         'atan(col_1 + 1)',
         'arctan(col_1 + 1)',
         'arctan2(col_1 + 1)',
         '- 1 * col_1+2 -(3*col_2)/(col_1) ', 
         'sin(col_2+2) -(col_1 + 1)',
         'abs(col_1 - 2)',
         'exp(-col_1)')
```

```python
results = []
for rule in rules:
    tree = from_string(rule)
    serialized = tree.serialize()
    results.append({'input': rule, 'output': serialized})
```

```python
with open('./tests/test.json', mode='w') as f:
    json.dump(results, f, indent=4)
```
